## Ray Tree



<b>Exercise</b>: if we have a ray tree with 1 R and 1 T branches, with colors R = (0.1,0.2,0.1), T = (0.2,0.0,0.1), Kr = 0.4 (Kr is reflection coefficient), Kt = 0.2 (Kt is transmittion coefficient), and P = (0.0,0.2,0.1) (P is the color of intersection point); what is the color of the pixel? <br>
- Color of pixel = $P + Kr * R + Kt * T$, for each color component
- What if we have more branches?

<b>Answer</b>:
- Compute for each color (RGB), e.g., $Red =  P + Kr * R + Kt * T = 0.0 + 0.1 * 0.4 + 0.2 * 0.2 = 0.08$
- If have more branches, just evaluate colors of each intersection from a bottom-up manner


## Pre-multiplied compositing color

<b>Exercise</b>: At a particular pixel, below are pre-multiplied colors and depth of 2 particles. Find the effective composited color (RGBA) in the following two cases: <br>
- Particle 1 color: (0.1, 0.2, 0.0, 0.2), Depth: 1.0 <br>
Particle 2 color: (0.1, 0.0, 0.1, 0.4), Depth: 2.0 <br>
- Particle 1 color: (0.0, 0.1, 0.2, 0.5), Depth: 2.0 <br>
Particle 2 color: (0.1, 0.3, 0.0, 1.0), Depth: 1.0 


<b>Answer</b>:
- Since $Depth_{P1} < Depth_{P2}$, particle 1 is at front, we compute color of the pixel this way: <br>
$Red = C_f + C_b * (1 - \alpha_f) = 0.1 + 0.1 * (1 - 0.2) = 0.18$<br>
$Green = C_f + C_b * (1 - \alpha_f) = 0.2 + 0.0 * (1 - 0.2) = 0.2$<br>
$Blue = C_f + C_b * (1 - \alpha_f) = 0.0 + 0.1 * (1 - 0.2) = 0.08$<br>
$Alpha = \alpha_f + \alpha_b * (1 - \alpha_f) = 0.2 + 0.4 * (1 - 0.2) = 0.52$<br><br>
- Since Particle 2 is at front ($Depth_{P2} < Depth_{P1}$) and its opacity ($\alpha$) value is 1.0, nothing behind it will be visible. <br>
Thus the color of the pixel is: $RGBA_{pixel} = RGBA_{P2} = (0.1, 0.3, 0.0, 1.0)$
