# Review of Lecture Content

## Parametric representation of line
- If P is a point on the line $P_1 P_2$, then express the point P using: $P = \alpha_1 * P_1 + \alpha_2 * P_2 $
- Useful link: https://tutorial.math.lamar.edu/classes/calciii/eqnsoflines.aspx


## Affine function
- An <b>affine function </b> v defined on points in a homogeneous coordinate system is a function that maps a point in homogeneous coordinates to a real number

## Affine combinition
- In a linear combinition of scalars/points/vectors where all coefficients sum to 1
- For a linear function such as $\alpha_1 * X_1 + \alpha_2 * X_2 + ... + \alpha_n * X_n$ where $X_i$ can be scalars/points/vectors, $\sum _{i=1} ^{n}{a_i} = 1$

## Convex combinition
- In a linear combinition of scalars/points/vectors where all coefficients are non-negative and sum to 1
- Built on affine combinition
- For a linear function such as $\alpha_1 * X_1 + \alpha_2 * X_2 + ... + \alpha_n * X_n$ where $X_i$ can be scalars/points/vectors, $\sum _{i=1} ^{n}{a_i} = 1$ and $\alpha_i \leq 0$

## Cross Product
- For two vectors $a$ and $b$, their cross product $a × b = |a| |b| sin(θ)$
- Useful link: https://www.mathsisfun.com/algebra/vectors-cross-product.html
